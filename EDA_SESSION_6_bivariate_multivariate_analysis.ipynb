- implement the covariance function

- covariance vs correlation

    - covariance will tell about the columns have relation or not

    - covariance will explain about relation status

        - postive relation

        - negative relation 

        - no relation

    - correlation tells about how much relation between two variables

    - it denotes with 'r'

    - r variance from -1 to 1

    - -1 to 0 indicates negative relation quantity

    - 0 to 1 indicates positive relation quantity

    - o indicates no relation

**method name : corr**

```python
visa_df.corr(numeric_only=True)
```

**Heat map**

- Heat map is one of the import visualization methpd , to show the matrix from 

- it is under **seaborn** package

- in every matrix we ave heighes values and lowest values avaialble

- Heat map will give colour visulization

- also it provides colour bar, which colour indicates which value

```python
corr_data=visa_df.corr(numeric_only=True)
sns.heatmap(corr_data,annot=True)
plt.show()
```

```python
corr_data=visa_df.corr(numeric_only=True)
sns.heatmap(corr_data,annot=True,cmap='Greens')
plt.show()
```
