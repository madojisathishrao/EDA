**finding the outliers**

- we already know that outliers available less than Q1-1.5*IQR and Q3+1.5*IQR

- step-1:

    - calculate Q1 = 25p
 
    - calculate Q2 = 50p
 
    - calculate Q3 = 75p

- step-2:

    - calculate IQR = Q3-Q1
 
- step-3:

    - calculte lb= Q1-1.5*IQR
 
    - calculate ub= Q3+1.5*IQR
 
- step-4:

    - con1= wage_data<lb
 
    - con2= wage_data>ub
 
    - con3=con1 and con2
 
- step-5:

- Data[con3]

```python
import numpy as np
import pandas as pd
data = {'wage_data': [10, 12, 14, 15, 18, 21, 22, 24, 30, 35, 40, 45, 50]}
df = pd.DataFrame(data)
Q1 = df['wage_data'].quantile(0.25)
Q2 = df['wage_data'].quantile(0.50)
Q3 = df['wage_data'].quantile(0.75)
IQR = Q3 - Q1
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR
con1 = df['wage_data'] < lb
con2 = df['wage_data'] > ub
con3 = con1 | con2
outliers = df[con3]
print("Lower Bound:", lb)
print("Upper Bound:", ub)
print("Outliers:\n", outliers)
```

```python
dict1={'sub':['DS','DE','CS'],
       'marks':[100,200,300]}
df=pd.DataFrame(dict1)
df

l=[]
for i in df.marks:
    if i>100:
        l.append(1)
    else:
        l.append(i)

df
```
