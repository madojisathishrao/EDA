**no.Where**

- using np.where also we can chage the categorical data to numerical data

- but np.where is only for binary condition

- np.where is same as if-else

- if condition is True : will apply True value

- if condition is False : will apply False value

- so np.where use only for binary lables

- a columns should have only two unique lables

```python
# value === 'certified' replace with 0
# otherwise 1
# np. where (con,0,1)
visa_df=pd.read_csv(r"C:\Users\sathish\Documents\datafiles\visadataset.csv")
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```
